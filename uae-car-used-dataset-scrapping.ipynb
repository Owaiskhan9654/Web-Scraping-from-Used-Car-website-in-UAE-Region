# **<span style="color:#9f51e7;">Some basics Import</span>**

```python
from bs4 import BeautifulSoup
import cloudscraper
from tqdm .notebook import tqdm
import pandas as pd
```

```python
%%capture

scraper = cloudscraper.create_scraper( browser={
        'browser': 'firefox',
        'platform': 'windows',
        'mobile': False
    }) # Cloud Scrapper Object
print(scraper.get("https://www.drivearabia.com/carprices/uae/embed/brandCountry/").text)
```

```python
content=scraper.get("https://www.drivearabia.com/carprices/uae/").text
soup = BeautifulSoup(content)
brands=soup.find_all({"a":{"class":{"brand-item"}}})
```

```python
carbrands_slug=[]
for i in brands:
    if "brand-item" in str(i):
        print(i.attrs['href'])
        carbrands_slug.append(i.attrs['href'])
        
```

```python
%%time

Website_links_all=[]
Website_links=[]
sub_brands_all=[]
sub_brands=[]


for i in tqdm(carbrands_slug):
    scraper = cloudscraper.create_scraper(  browser={
        'browser': 'firefox',
        'platform': 'windows',
        'mobile': False
    }) 
    url="{}".format(i)
    content=scraper.get(url).text
    soup = BeautifulSoup(content)
    brands=soup.find("div", {"class": "featured-title"})
    brands_link=soup.find_all("h2", {"class": "featured-title"})
    for i in brands_link:
        b=i.find('a',href=True)
        Website_links.append(b['href'])
        name=b.contents
        name=name[0].replace('\n','')
        name= name.replace(' ','')
        sub_brands.append(name)
for k in tqdm(Website_links):
    scraper = cloudscraper.create_scraper(  browser={'browser': 'firefox','platform': 'windows','mobile': False})
    content3=scraper.get(k).text
    soup3 = BeautifulSoup(content3)
    brands_link=soup3.find_all("div", {"class": "row one-pixel-gutter"})
    Models=[]
    for i in brands_link:
        b=i.find_all('a',href=True)
        Website_links_model=[]
        for j in range(0,len(b)):
            #print(b[j]['href'])
            Website_links_model.append(b[j]['href'])
            name=b[j].contents
            name=name[0].replace('\n','')
            name= name.replace(' ','')
            Models.extend(name)
    sub_brands_all.extend(Models)
    Website_links_model=list(set(Website_links_model))
    Website_links_all.extend(Website_links_model)

        

Website_links_all[0:15]
```

```python
len(list(set(Website_links_all)))
```

```python
Website_links_all = [i for i in Website_links_all if i != 'javascript:;']
```

```python
Website_links_all.count('javascript:;')
```

<center><h1 style = "font-size:25px;font-family: Comic Sans MS">Please remove the condition of break loop if you want to scrape all the data locally</h1></center>

```python
origin_country=[]
Approx_cost=[]
BodyType=[]
additionalType=[]
weight=[]
Engine=[]
Gearbox=[]
Power=[]
Torque=[]
Fuel_Econ_L_100km=[]
Fuel_Econ_km_L=[]
sale_country=[]
Performance_0_100=[]
Top_speed_kph=[]
Overview=[]

Manufacturer=[]
Brand=[]
Model=[]

j=0
for i in tqdm(Website_links_all):
    scraper = cloudscraper.create_scraper(  browser={
        'browser': 'firefox',
        'platform': 'windows',
        'mobile': False
    }) 
    content=scraper.get(i).text
    soup = BeautifulSoup(content, 'html.parser')

    try:
        Approx_cost.append(soup.find("td",{'class':'no-rtl'}).contents[0])
    except:
        Approx_cost.append('NA')
        
    try:
        origin_country.append(soup.find("td",{'dir':'ltr'}).contents[0])
    except:
        origin_country.append('NA')       
        
    try:
        BodyType.append(soup.find("td",{'itemprop':'bodyType'}).contents[0])
    except:
        BodyType.append('NA')
        
    try:
        additionalType.append(soup.find("td",{'dir':'ltr','itemprop':"additionalType"}).contents[0])
    except:
        additionalType.append('NA')
        
    try:
        weight.append(soup.find("td",{'dir':'ltr','itemprop':"weight"}).contents[0])
    except: 
        weight.append('NA')   
    
    try:
        sale_country.append(i[i.index('carprices/')+len('carprices/'):i.find('/',i.index('carprices/')+len('carprices/'))])
    except:
        sale_country.append('NA')        
    
    try:
        Engine.append(soup.find("span",{'itemprop':"name"}).contents[0])
    except:
        Engine.append('NA')
        
    try:
        Gearbox.append(soup.find("td",{'itemprop':"vehicleTransmission"}).contents[0])
    except:
        Gearbox.append('NA')
        
    try:
        Power.append(soup.find("span",{'itemprop':"enginePower"}).contents[0])
    except:
        Power.append('NA')
        
    try:
        Torque.append(soup.find("span",{'itemprop':"torque"}).contents[0])
    except:
        Torque.append('NA')
        
    try:
        Fuel_Econ_L_100km.append(soup.find("td",{'itemprop':"fuelEfficiency"}).contents[0])
    except:
        Fuel_Econ_L_100km.append('NA')
        
    try:
        Fuel_Econ_km_L.append(soup.find_all("td",{'itemprop':"fuelEfficiency"})[1].contents[0])
    except:
        Fuel_Econ_km_L.append('NA')
    
    try:
        Performance_0_100.append(soup.find("td",{'itemprop':"speed"}).contents[0])
    except:
        Performance_0_100.append('NA')
    
    try:
        Top_speed_kph.append(soup.find_all("td",{'itemprop':"speed"})[1].contents[0])
    except:    
        Top_speed_kph.append('NA')
    
    try:
        Overview.append(soup.find("div",{'class':"vehicle-overview"}).text.replace('\n',''))
    except:
        Overview.append('NA')
        
    try:
        Manufacturer.append(soup.find("span",{'itemprop':"manufacturer"}).contents[0])
    except:
        Manufacturer.append('NA')
        
    try:
        Brand.append(soup.find("span",{'itemprop':"brand"}).contents[0])
    except:
        Brand.append('NA')
        
    try:
        Model.append(soup.find("span",{'itemprop':"model"}).contents[0])
    except:
        Model.append('NA')
        
    j=j+1
    if j==2000:
        break
```

# **<span style="color:#9f51e7;">Parsing all the data into DataFrame and then exporting to csv</span>**

```python
df=pd.DataFrame({'Link':Website_links_all[0:2000],'Approx Cost':Approx_cost,'Origin Country':origin_country,\
                 'Sale Country':sale_country,'Manufacturer':Manufacturer,'Brand':Brand,'Model Year':Model,\
                 'Body Type':BodyType,'Additional Type':additionalType,\
                 'Weight':weight,'Gear box':Gearbox,'Power (hp)':Power,'Torque (Nm)':Torque,\
                 'Fuel Econ (L/100km)':Fuel_Econ_L_100km,'Fuel Econ (km/L)':Fuel_Econ_km_L,\
                 'Performance 0-100 kph (sec)':Performance_0_100,'Top speed (kph)':Top_speed_kph,\
                 'Overview':Overview})
```
